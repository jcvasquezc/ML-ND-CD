# Unemployment dataset

Firms (firms) demand different factors to produce goods or services. One of those factors is work. For their part, households offer their labor in exchange for a wage that allows them to acquire goods or services in the economy. Several aspects, both short and long term, do not allow at a specific time all firms find the amount of work they are demanding or that all people who offer their work capacity are employed.

Employment and unemployment rates, calculated by the National Administrative Department of Statistics (** DANE **), are indicators of the evolution over time of, respectively, the proportion of people who, being of working age, are employed and Of the proportion of people who, intending to work, can be employed.



### Explanatory notes:

* The employment rate (occupancy rate) refers to the percentage ratio between the employed population and the working-age population.
* The unemployment rate is the percentage ratio between the number of unemployed persons and the economically active population.
* For the calculations the following thirteen cities are considered: Bogotá, Cali, Medellín, Barranquilla, Bucaramanga, Manizales, Pasto, Pereira, Cúcuta, Villavicencio, Ibagué, Monteria and Cartagena.
* Projections calculated in 2012: calculated figures with population projections based on the 2005 census updated in January 2012. 

### Metadata
Information since January 2001 until May 2017 with monthly resolution.

## Matrix structure

```python
# Cast data as matrix structure
TDnp = np.matrix(TD)

# Creating matrix X and vector Y
[X , Y]= [TDnp[:,0],TDnp[:,3]]
```

## Linear model Creation

```python
# intanciando Linear model
lm = linear_model.LinearRegression()

# Train linear model
lm.fit(X,Y)

# Extract score
original_score = '{0:.3f}'.format( lm.score( X, Y ) )

# Predict values
predictedValues = lm.predict(TDnp[:,0])
```
